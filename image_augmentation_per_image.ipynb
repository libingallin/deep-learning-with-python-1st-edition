```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing import image

%matplotlib inline
```

```python
print('TensorFlow version:', tf.__version__)
```

```python
image_path = 'figs/chap05-figs/cat.1700.jpg'
# Loads an image into PIL format.
image_pil = image.load_img(path=image_path, color_mode='rgb',
                           target_size=None, interpolation='nearest')
image_pil
```

```python
# Converts a PIL Image instance to a Numpy array.
image_arr = image.img_to_array(image_pil)
image_arr.shape, image_arr.dtype
```

```python
plt.imshow(image_arr.astype('uint8'))
plt.show()
```

```python
# Converts a 3D Numpy array to a PIL Image instance.
image.array_to_img(image_arr)
```

```python
plt.imshow(image.array_to_img(image_arr))
plt.show()
```

```python
datagen = image.ImageDataGenerator(
    # int. Degree range for random rotations.
    # 表示图像随机旋转的角度范围
    rotation_range=90)

i = 0
plt.figure(figsize=(14, 3))
for image_arr_generated in datagen.flow(image_arr.reshape((-1, )+image_arr.shape)):
    plt.subplot(1, 4, i+1)
    # image_arr_generated shape = (1, 333, 324, 3)
    plt.imshow(image.array_to_img(image_arr_generated[0]))
    i += 1
    if i == 4:
        break
plt.show()
```
